```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mask_path = "../../Field Data/Individual Anatomy/Galaxy M34/Left Palm_files/left_palm_segmasks/SegmentationClass/1709875760020.png"

img_path = "../../Field Data/Individual Anatomy/Galaxy M34/Left Palm_files/Images_left_palm/1709875760020.jpg"
```

```python
mask_bgr = cv2.imread(mask_path)
img_bgr = cv2.imread(img_path)
```

```python
mask = cv2.cvtColor(mask_bgr,cv2.COLOR_BGR2GRAY)
img = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(mask)
```

```python
plt.imshow(img)
```

```python
dest_and = cv2.bitwise_and(img, img, mask = mask)
plt.imshow(dest_and)
```
